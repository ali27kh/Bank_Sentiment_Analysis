```python
import torch
from transformers import PegasusForConditionalGeneration, PegasusTokenizer
model_name = 'tuner007/pegasus_paraphrase'
torch_device = 'cuda' if torch.cuda.is_available() else 'cpu'
tokenizer = PegasusTokenizer.from_pretrained(model_name)
model = PegasusForConditionalGeneration.from_pretrained(model_name).to(torch_device)

def get_response(input_text,num_return_sequences,num_beams):
    batch = tokenizer([input_text],truncation=True,padding='longest',max_length=60, return_tensors="pt").to(torch_device)
    translated = model.generate(**batch,max_length=60,num_beams=num_beams, num_return_sequences=num_return_sequences, temperature=1.5)
    tgt_text = tokenizer.batch_decode(translated, skip_special_tokens=True)
    return tgt_text
```

```python
num_beams = 10
num_return_sequences = 10
```

```python
response=get_response("the most wonderful bank, excellent treatment ",num_return_sequences,num_beams)
response
```

```python
import pandas as pd
posdata=pd.read_csv('positive_sa.csv')
negdata=pd.read_csv('negative_sa.csv')
```

```python
posdata
```

```python
# commentaire to list
pos_reviews=posdata['review'].tolist()
neg_reviews=posdata['review'].tolist()
```

```python
for i in range (len(pos_reviews)):
  context=pos_reviews[i]
  response=get_response(context,num_return_sequences,num_beams)
  for j in response :
    pos_reviews.append(j)

for i in range (len(neg_reviews)):
  context=neg_reviews[i]
  response=get_response(context,num_return_sequences,num_beams)
  for j in response :
    neg_reviews.append(j)
```

```python
# neg_reviews
```

```python
dfpositive = pd.DataFrame(pos_reviews, columns=['review'])
dfpositive['sentiment']='positive'
dfnegative = pd.DataFrame(neg_reviews, columns=['review'])
dfnegative['sentiment']='negative'
```

```python
df=pd.concat([dfpositive,dfnegative], axis=0)
```
